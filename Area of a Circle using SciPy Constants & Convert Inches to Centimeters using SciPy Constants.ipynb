```python
import pandas as pd
import scipy
from scipy import constants as cons
```

```python
csv_file="https://raw.githubusercontent.com/AnudipAE/DANLC/master/radius_data.csv"

#read CSV File
d_frame=pd.read_csv(csv_file)

#calculate Area and add a column
d_frame["Area"]=d_frame["Radius"]*d_frame["Radius"]*cons.pi

#show top 5 data
d_frame.head(5)
```

```python
# Save to a CSV FILE
d_frame.to_csv("https://raw.githubusercontent.com/EtikaSingh/Python_Anudip/main/Area_data.csv",index=False)
```

```python
import pandas as pd
import scipy
from scipy import constants as cons
```

```python
csv_file="https://raw.githubusercontent.com/AnudipAE/DANLC/master/people_heights.csv"

#read CSV File
d_frame=pd.read_csv(csv_file)

#calculate height in cm and add a column
d_frame["Height (cm)"]=d_frame["Height (inches)"]*cons.inch*100

#show top 5 data
d_frame.head(5)
```

```python
# Save to a CSV FILE
d_frame.to_csv("https://raw.githubusercontent.com/EtikaSingh/Python_Anudip/main/Area_data.csv",index=False)
```
