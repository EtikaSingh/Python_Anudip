```python
import pandas as pd
import matplotlib.pyplot as plt
csv_file_path ="https://raw.githubusercontent.com/EtikaSingh/Python_Anudip/main/Cleaned_salesdata.csv"

#read the csv file
d_frame=pd.read_csv(csv_file_path)

#create pivot table
pd.pivot_table(d_frame,index=["Region","Manager","SalesMan"],values=["Sale_amt"],aggfunc=["sum"])
```

```python
#import pandas library
import pandas as pd

csv_file_path="https://raw.githubusercontent.com/EtikaSingh/Python_Anudip/main/Cleaned_salesdata.csv"

#read the csv file
d_frame=pd.read_csv(csv_file_path)

#create the pivot table
pd.pivot_table(d_frame,index=["Item"],values=["Units"],aggfunc=["sum"])
```
